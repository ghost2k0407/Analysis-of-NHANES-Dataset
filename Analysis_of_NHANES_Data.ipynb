### **3.Compute Basic Numerical Aggregates for Male and Female Weights**

```python
# Function to compute statistics
def compute_statistics(data):
    mean = np.mean(data)
    median = np.median(data)
    variance = np.var(data, ddof=1)  # Sample variance
    std_dev = np.std(data, ddof=1)  # Sample standard deviation
    data_range = np.max(data) - np.min(data)
    skewness = skew(data)
    kurt = kurtosis(data)
    
    return mean, median, variance, std_dev, data_range, skewness, kurt

# Compute statistics for female weights
mean_f, median_f, variance_f, std_dev_f, range_f, skewness_f, kurt_f = compute_statistics(arr1[:, 0])

# Compute statistics for male weights
mean_m, median_m, variance_m, std_dev_m, range_m, skewness_m, kurt_m = compute_statistics(arr2[:, 0])

# Print results
print(f"Female Weights:\n Mean: {mean_f:.2f}, Median: {median_f:.2f}, Variance: {variance_f:.2f}, Std Dev: {std_dev_f:.2f}, Range: {range_f:.2f}, Skewness: {skewness_f:.2f}, Kurtosis: {kurt_f:.2f}")
print(f"Male Weights:\n Mean: {mean_m:.2f}, Median: {median_m:.2f}, Variance: {variance_m:.2f}, Std Dev: {std_dev_m:.2f}, Range: {range_m:.2f}, Skewness: {skewness_m:.2f}, Kurtosis: {kurt_m:.2f}")
```

### **4.Find Body Mass Index for The Female Matrix and Add it as Column**

```python
# Assuming the second column represents height in meters
heights = arr1[:, 1]
weights = arr1[:, 0]

# Calculate BMI = weight (kg) / height (m)^2
bmi = weights / (heights ** 2)

# Add BMI as the eighth column to the female matrix
arr = np.column_stack((arr1, bmi))

print(np.round(arr[:5],2))  # Display the first 5 rows to verify the result
```

### **5.Find the Standardized Matrix for Female Matrix Data**

```python
# Calculate the mean and standard deviation for each column
means = np.mean(arr, axis=0)
std_devs = np.std(arr, axis=0)

# Standardize each column
zfemale = (arr - means) / std_devs

# Display the first 5 rows of the standardized matrix
print(np.round(zfemale[:5],2))
```

### **9.Advantages and Disadvantages of BMI,Waist to Height ratio,Waist to Hip ratio**

### **BMI (Body Mass Index)**

**Advantages:**

Simple to calculate using height and weight.

Widely used and recognized as a general indicator of health.

Can identify potential health risks associated with being underweight or overweight.

**Disadvantages:**

Does not differentiate between muscle mass and fat.

Not accurate for athletes or people with muscular builds.

Does not account for body fat distribution.


### **Waist-to-Height Ratio (WHtR)**

**Advantages:**

Easier to measure than waist-to-hip ratio.

Reflects abdominal fat distribution, which is linked to health risks.

May be a better predictor of health risks than BMI for some populations.

**Disadvantages:**

Less widely used than BMI.

May not be as accurate as waist-to-hip ratio in assessing body fat distribution.

Still doesn't account for muscle mass or overall body composition.


### **Waist-to-Hip Ratio (WHR)**

**Advantages:**

Better indicator of abdominal fat distribution than BMI.

Strongly linked to health risks like heart disease, diabetes, and stroke.

Can be used to assess overall body shape and health.

**Disadvantages:**

More complex to measure than BMI or WHtR.

Requires accurate measurements of waist and hip circumference.

May not be as reliable for people with very large or small builds.
